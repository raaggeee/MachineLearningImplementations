## Perceptron Trick
Perceptron Trick is a way to adjust the weights and bias using the misclassified points to minimize the error.\
Consider we have a 2D dataset. Here, we update the weights of line Ax + By + C = 0 - A, B and intercept C to get a line which classifies the binary dataset.


### When A is changed, the line is rotated along x-axis
### When B is changed, the line is rotated along y-axis
### When C is changed, the line moves paralelly




## Steps for algorithm
1. for i -> len(X):\
2.      select a random row
3.      make a new prediction with step function
4.      then update the weight

# Implementing perceptron trick

```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0,
                           random_state=42, n_classes=2, n_clusters_per_class=1, hypercube=False, class_sep=10)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="winter", s=100)
```

```python
def step(z):
    return 1 if z > 0 else 0

def perceptron(X, y, lr):
    X = np.insert(X, 0, 1, axis=1)
    W = np.ones(X.shape[1])

    for i in range(X.shape[0]):
        j = np.random.randint(0, 100)
        y_hat = step(np.dot(X[j], W))
        W = W + lr * (y[j] - y_hat) * X[j]

    return W[0], W[1:]
```

```python
intercept_, coeff_ = perceptron(X, y, 0.01)
```

```python
print(intercept_)
print(coeff_)
```
