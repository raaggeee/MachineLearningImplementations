## Regression Trees
Regression Trees are decision trees which are used to predict regression value. It works well with non-linear data.

Consider we have a 2D dataset of average walking speed and distance covered.



In this case what regression tree does that it creates splits of data based on features. 



Then it predicts the value based on mean value of all the points present in that particular region. 

Ex: If we have to predict how much will a person cover distance if avg speed is 4m/s. In that case we find avg of all the points between 3 - 5 m/s.
